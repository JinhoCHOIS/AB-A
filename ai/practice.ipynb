```python
import pandas as pd
import scipy as sc
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.proportion import proportions_ztest
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sc.random.seed(seed=1234)
normal=sc.random.normal(20, 2, 100000)
sns.distplot(normal, fit=stats.norm, kde=False)
normal.round(3)
```

```python
sc.random.seed(seed=1234)
normal2=sc.random.normal(0,1,100000)
sns.distplot(normal2, fit=stats.chi, kde=True)
normal2.round(3)
```

```python
mu = 115
sigma = 20
x = 135
prop = stats.norm.cdf(x, mu, sigma)
print("{0}분 이상 걸리는 A/S작업 비율 : {1:.1f}%". format(x, (1-prop)*100))
```

```python
mu = 175
sigma = 5
x = 180
prop = stats.norm.cdf(x, mu, sigma)
print("{0}cm면 상위 {1:.1f}%에 속합니다.". format(x, (1-prop)*100))
```
