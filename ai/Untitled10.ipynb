# 어떤 자동차 1리터당 주행거리가 평균 12km, 표준편차가 3km인 정규분포를 이룬다고 하자
# 1)1리터를 가지고 12km이상 15km 이하를 달릴 확률은?
# 2)1리터를 가지고 9km이상 15km 이하를 달릴 확률은?
# 3)1리터를 가지고 15km이상을 달릴 확률은?

```python
mu = 12
sigma = 3
x1 = 15
x2 = 9

p1 = stats.norm.cdf(x1, mu, sigma)
p2 = stats.norm.cdf(x2, mu, sigma)

print('1)1리터를 가지고 {0}km이상 ,{1}km이하를 달릴 확률은 {2:.1f}%입니다.'.format(mu, x1, (p1-0.5)*100))
print('2)1리터를 가지고 {0}km이상 ,{1}km이하를 달릴 확률은 {2:.1f}%입니다.'.format(x2, x1, (p1-p2)*100))
print('3)1리터를 가지고 {0}km이상 달릴 확률은 {1:.1f}%입니다.'.format(x1, (1-p1)*100))
```

```python
mu = 2
prop = stats.poisson.pmf(3, mu)
cdf_prop = stats.poisson.cdf(2, mu)
print('1분당 {0}번의 전화가 걸려올 확률 : {1:.5f}'.format(3, prop))
print('1분당 최대 {0}회의 이하의 전화가 걸려올 확률 : {1:.5f}'.format(3, cdf_prop))
```

```python
mu = 2
prop = stats.poisson.pmf(2, mu)
cdf_prop = stats.poisson.cdf(1, mu)
cdf_prop2 = stats.poisson.cdf(2, mu)

print('{0}명의 손님이 도착할 확률 : {1:.5f}'.format(2, prop))
print('{0}명 이하의 손님이 도착할 확률 : {1:.5f}'.format(1, cdf_prop))
print('{0}명 이상의 손님이 도착할 확률 : {1:.5f}'.format(2, 1-cdf_prop2))
```
