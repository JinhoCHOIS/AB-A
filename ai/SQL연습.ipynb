```python
import pandas as pd
```

```python
import sqlite3
```

```python
connection = sqlite3.connect('DB - Test.db')
```

```python
query = "SELECT * FROM sqlite_master;"
```

```python
pd.read_sql_query(query,connection)['sql']
```

```python
query = "SELECT * FROM 'user'"
user = pd.read_sql(query, connection)
print(user.shape)
user.head()
```

```python
query = "SELECT * FROM 'log' "
log = pd.read_sql(query, connection)
print(log.shape)
log.head()
```

# MERGE

```python
user.shape
```

```python
log.shape
```

```python
pd.merge(user, log).head() #중복되지 않은 data는 제거
```

```python
pd.merge(user, log, how='outer') #중복 허용
```

```python
t1=pd.merge(user,log,how='inner')
t1.shape
```

```python
t2=pd.merge(user,log,how='outer')
t2.shape
```

```python
t3=pd.merge(user,log,how='left')
t3.shape
```
