```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import f1_score
```

```python
ds_ch = pd.read_csv('./Downloads/week4day1/통신고객이탈.csv',engine='python')
ds_ch.info()
```

```python
ds_sel=ds_ch.select_dtypes(exclude='object')
ds_sel.info()
```

```python
ds_ch_dummy=pd.get_dummies(ds_ch)
ds_ch_dummy.head()
```

```python
ds_ch["CHURN"].replace(to_replace = "Active", value=0, inplace=True)
ds_ch["CHURN"].replace(to_replace = "Churned", value=1, inplace=True)
ds_sel["CHURN"]=ds_ch["CHURN"]
ds_ch_dummy=pd.get_dummies(ds_sel)
ds_ch_dummy.head()
```

```python
ds_ch_scaled = scale(ds_ch_dummy)
ds_ch_scaled = pd.DataFrame(ds_ch_scaled, columns = ds_ch_dummy.columns)
ds_ch_scaled["CHURN"] = ds_ch["CHURN"]
ds_ch_scaled.head()
```

```python
np.random.seed(seed=1234)
msk=np.random.rand(ds_ch_dummy.shape[0]) <0.7
ds_ch_train=ds_ch_dummy[msk]
ds_ch_test=ds_ch_dummy[~msk]

ds_ch_train_y=ds_ch_train["CHURN"]
ds_ch_train_x=ds_ch_train.drop("CHURN", axis=1, inplace=False)
ds_ch_test_y=ds_ch_test["CHURN"]
ds_ch_test_x=ds_ch_test.drop("CHURN", axis=1, inplace=False)

print("train data X size : {}".format(ds_ch_train_x.shape))
print("train data Y size : {}".format(ds_ch_train_y.shape))
print("test data X size : {}".format(ds_ch_test_x.shape))
print("test data Y size : {}".format(ds_ch_test_y.shape))
```

```python
ds_ch_train_x_scaled=scale(ds_ch_train_x, axis=0)
ds_ch_test_x_scaled=scale(ds_ch_test_x, axis=0)
pd.DataFrame(ds_ch_train_x_scaled, columns=ds_ch_train_x.columns).head()
```

```python
model=["DecisionTree", "RandomForest", "GradientBoosting", "SVM"]
train_accuracy=[]
test_accuracy=[]
model_f1_score=[]
```

```python
tree = DecisionTreeClassifier(criterion="gini", max_depth=10, min_samples_leaf=30,random_state=1234)
tree.fit(ds_ch_train_x_scaled, ds_ch_train_y)
train_accuracy.append(tree.score(ds_ch_train_x_scaled, ds_ch_train_y))
test_accuracy.append(tree.score(ds_ch_test_x_scaled, ds_ch_test_y))
model_f1_score.append(f1_score(ds_ch_test_y,tree.predict(ds_ch_test_x_scaled)))
```

```python
rf = RandomForestClassifier(criterion="gini", n_estimators=100, max_depth=4, min_samples_leaf=30, random_state=1234)
rf.fit(ds_ch_train_x_scaled, ds_ch_train_y)
train_accuracy.append(rf.score(ds_ch_train_x_scaled, ds_ch_train_y))
test_accuracy.append(rf.score(ds_ch_test_x_scaled, ds_ch_test_y))
model_f1_score.append(f1_score(ds_ch_test_y,rf.predict(ds_ch_test_x_scaled)))
```

```python
gb = GradientBoostingClassifier(learning_rate=0.1, max_depth=4, min_samples_leaf=30, n_estimators=5, random_state=1234)
gb.fit(ds_ch_train_x_scaled, ds_ch_train_y)
train_accuracy.append(gb.score(ds_ch_train_x_scaled, ds_ch_train_y))
test_accuracy.append(gb.score(ds_ch_test_x_scaled, ds_ch_test_y))
model_f1_score.append(f1_score(ds_ch_test_y,gb.predict(ds_ch_test_x_scaled)))
```

```python
svm = SVC(C=6, gamma=0.1, random_state=1234)
svm.fit(ds_ch_train_x_scaled, ds_ch_train_y)
train_accuracy.append(svm.score(ds_ch_train_x_scaled, ds_ch_train_y))
test_accuracy.append(svm.score(ds_ch_test_x_scaled, ds_ch_test_y))
model_f1_score.append(f1_score(ds_ch_test_y,svm.predict(ds_ch_test_x_scaled)))
```

```python
ds_eval=pd.DataFrame(index=model)
ds_eval["TrainAccuracy"]=train_accuracy
ds_eval["TestAccuracy"]=test_accuracy
ds_eval["F1Score"]=model_f1_score
ds_eval.round(3)
```
